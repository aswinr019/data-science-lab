```python
import numpy as np
import pandas as pd
```

```python
house_price = pd.read_csv('/content/drive/My Drive/houseprice.csv')
```

```python
house_price.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder , StandardScaler
label_encoder = LabelEncoder()
```

```python
fields = ['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus']
```

```python
for field in fields:
  house_price[field]  = label_encoder.fit_transform(house_price[field])
```

```python
house_price.head(20)
```

```python
X = house_price.drop('price',axis=1)
y = house_price['price']
```

```python
sc = StandardScaler()
X_scaled = sc.fit_transform(X)

print(X_scaled)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train, y_test = train_test_split(X_scaled,y,test_size=0.2,random_state=12)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
print(y_predict)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_predict)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_predict)
print("Mean Squared Error (MSE):",mse)
print("Root Mean Squared Error (RMSE):",rmse)
print("R-squared (R2):",r2)
```
